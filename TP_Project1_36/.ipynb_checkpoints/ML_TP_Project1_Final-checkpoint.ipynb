# Imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score, matthews_corrcoef, confusion_matrix, accuracy_score
from sklearn.metrics import explained_variance_score, mean_squared_error, max_error, mean_absolute_error
from scipy.stats import pearsonr
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

# Dataset

### Dependent variables:
* 'total UPRDS' 
* 'motor UPRDS' 

```python
df = pd.read_csv('parkinsons_updrs.data')
df
```

```python
# it was ran once just to select important variables
excluded_columns = ['total_UPDRS','motor_UPDRS', 'subject#', 'age', 'sex', 'test_time']
X = df.loc[:, ~df.columns.isin(excluded_columns)].to_numpy()
y = df['motor_UPDRS'].to_numpy()
```

# Selecting crucial variables

```python
from statsmodels.api import OLS, add_constant
X_tr = add_constant(X)
mdl=OLS(y, X_tr, hasconst=12).fit()
mdl.summary()
```

```python
important_columns = ['Jitter(Abs)', 'Shimmer:APQ5', 'Shimmer:APQ11', 'NHR', 'HNR', 'DFA', 'PPE']
X = df.loc[:, df.columns.isin(important_columns)].to_numpy()
y = df['motor_UPDRS'].to_numpy()
```

# Objective 1

## Regression model for 'motor UPDRS'

```python
X_train, X_IVS, y_train, y_IVS = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
import numpy as np
from sklearn.model_selection import KFold

kf = KFold(n_splits=10, shuffle=True, random_state=23)
kf.get_n_splits(X_train)

rve_values = []
rmse_values = []
correlation_values = []
max_values = []
mae_values = []

# min_samples_leaf
for i in range(1, 21):
    TRUTH_nfold=None
    PREDS_nfold=None
    for train_index, test_index in kf.split(X_train):
        Xk_train, Xk_test = X_train[train_index], X_train[test_index]
        yk_train, yk_test = y_train[train_index], y_train[test_index]

        mdl = DecisionTreeRegressor(min_samples_leaf=i, max_depth=5, random_state=3)
        mdl.fit(Xk_train, yk_train)
        preds = mdl.predict(Xk_test)
        if TRUTH_nfold is None:
            PREDS_nfold=preds
            TRUTH_nfold=yk_test
        else:
            PREDS_nfold=np.hstack((PREDS_nfold, preds))
            TRUTH_nfold=np.hstack((TRUTH_nfold, yk_test))

    print(f"Statistics for {i} min_samples_leaf hyperparameter: \n--------------------------")
    print("The RVE is: ", explained_variance_score(TRUTH_nfold, PREDS_nfold))
    print("The rmse is: ", mean_squared_error(TRUTH_nfold, PREDS_nfold, squared=False))
    corr, pval=pearsonr(TRUTH_nfold, PREDS_nfold)
    print("The Correlation Score is is: %6.4f (p-value=%e)"%(corr,pval))
    print("The Maximum Error is is: ", max_error(TRUTH_nfold, PREDS_nfold))
    print("The Mean Absolute Error is: ", mean_absolute_error(TRUTH_nfold, PREDS_nfold))
    print("--------------------------")
    rve_values.append(explained_variance_score(TRUTH_nfold, PREDS_nfold))
    rmse_values.append(mean_squared_error(TRUTH_nfold, PREDS_nfold, squared=False))
    correlation_values.append((corr,pval))
    max_values.append(max_error(TRUTH_nfold, PREDS_nfold))
    mae_values.append(mean_absolute_error(TRUTH_nfold, PREDS_nfold))
    
# max_depth
for i in range(1, 21):
    TRUTH_nfold=None
    PREDS_nfold=None
    for train_index, test_index in kf.split(X_train):
        Xk_train, Xk_test = X_train[train_index], X_train[test_index]
        yk_train, yk_test = y_train[train_index], y_train[test_index]

        mdl = DecisionTreeRegressor(max_depth=i, random_state=3)
        mdl.fit(Xk_train, yk_train)
        preds = mdl.predict(Xk_test)
        if TRUTH_nfold is None:
            PREDS_nfold=preds
            TRUTH_nfold=yk_test
        else:
            PREDS_nfold=np.hstack((PREDS_nfold, preds))
            TRUTH_nfold=np.hstack((TRUTH_nfold, yk_test))
            
    print(f"Statistics for {i} max_depth hyperparameter: \n--------------------------")
    print("The RVE is: ", explained_variance_score(TRUTH_nfold, PREDS_nfold))
    print("The rmse is: ", mean_squared_error(TRUTH_nfold, PREDS_nfold, squared=False))
    corr, pval=pearsonr(TRUTH_nfold, PREDS_nfold)
    print("The Correlation Score is is: %6.4f (p-value=%e)"%(corr,pval))
    print("The Maximum Error is is: ", max_error(TRUTH_nfold, PREDS_nfold))
    print("The Mean Absolute Error is: ", mean_absolute_error(TRUTH_nfold, PREDS_nfold))
    print("--------------------------")
    rve_values.append(explained_variance_score(TRUTH_nfold, PREDS_nfold))
    rmse_values.append(mean_squared_error(TRUTH_nfold, PREDS_nfold, squared=False))
    correlation_values.append((corr,pval))
    max_values.append(max_error(TRUTH_nfold, PREDS_nfold))
    mae_values.append(mean_absolute_error(TRUTH_nfold, PREDS_nfold))
```

# Plots

```python
fig, plots = plt.subplots(nrows=1, ncols=5, figsize=(30, 10))

# min_samples_leaf
number_of_leaves = [i for i in range(1, 21)]
plots[0].set_xlabel("min_samples_leaf")
plots[0].set_ylabel("RVE")
plots[0].plot(number_of_leaves, rve_values[:20])
plots[1].set_xlabel("min_samples_leaf")
plots[1].set_ylabel("RMSE")
plots[1].plot(number_of_leaves, rmse_values[:20])
plots[2].set_xlabel("min_samples_leaf")
plots[2].set_ylabel("Correlation")
plots[2].plot(number_of_leaves, correlation_values[:20])
plots[3].set_xlabel("min_samples_leaf")
plots[3].set_ylabel("Max error")
plots[3].plot(number_of_leaves, max_values[:20])
plots[4].set_xlabel("min_samples_leaf")
plots[4].set_ylabel("MAE")
plots[4].plot(number_of_leaves, mae_values[:20])

fig, plots = plt.subplots(nrows=1, ncols=5, figsize=(30, 10))

# max_depth
plots[0].set_xlabel("max_depth")
plots[0].set_ylabel("RVE")
plots[0].plot(number_of_leaves, rve_values[20:])
plots[1].set_xlabel("max_depth")
plots[1].set_ylabel("RMSE")
plots[1].plot(number_of_leaves, rmse_values[20:])
plots[2].set_xlabel("max_depth")
plots[2].set_ylabel("Correlation")
plots[2].plot(number_of_leaves, correlation_values[20:])
plots[3].set_xlabel("max_depth")
plots[3].set_ylabel("Max error")
plots[3].plot(number_of_leaves, max_values[20:])
plots[4].set_xlabel("max_depth")
plots[4].set_ylabel("MAE")
plots[4].plot(number_of_leaves, mae_values[20:])
```

```python
# the best regression model is received by setting min_samples_leaf = 11 and max_depth = 5
TRUTH_nfold=None
PREDS_nfold=None
for train_index, test_index in kf.split(X_train):
    Xk_train, Xk_test = X_train[train_index], X_train[test_index]
    yk_train, yk_test = y_train[train_index], y_train[test_index]

    mdl = DecisionTreeRegressor(min_samples_leaf=17, max_depth=7, random_state=3)
    mdl.fit(Xk_train, yk_train)
    preds = mdl.predict(Xk_test)
    if TRUTH_nfold is None:
        PREDS_nfold=preds
        TRUTH_nfold=yk_test
    else:
        PREDS_nfold=np.hstack((PREDS_nfold, preds))
        TRUTH_nfold=np.hstack((TRUTH_nfold, yk_test))

# plt.figure(figsize=(15,20))
# r=plot_tree(mdl, filled=True)
# plt.show()

plt.figure(figsize=(5,5))
plt.scatter(PREDS_nfold, TRUTH_nfold)
plt.plot((0, 50), (0,50), c="r")
plt.grid()
plt.show()

print("Cross validation statistics for best parameters:")
print("The RVE is: ", explained_variance_score(TRUTH_nfold, PREDS_nfold))
print("The rmse is: ", mean_squared_error(TRUTH_nfold, PREDS_nfold, squared=False))
corr, pval=pearsonr(TRUTH_nfold, PREDS_nfold)
print("The Correlation Score is is: %6.4f (p-value=%e)"%(corr,pval))
print("The Maximum Error is is: ", max_error(TRUTH_nfold, PREDS_nfold))
print("The Mean Absolute Error is: ", mean_absolute_error(TRUTH_nfold, PREDS_nfold))

mdl.fit(X_train, y_train)
preds = mdl.predict(X_IVS)
print("-------------------------------")
print("The best model statistics are:")
print("The RVE is: ", explained_variance_score(y_IVS, preds))
print("The rmse is: ", mean_squared_error(y_IVS, preds, squared=False))
corr, pval=pearsonr(y_IVS, preds)
print("The Correlation Score is is: %6.4f (p-value=%e)"%(corr,pval))
print("The Maximum Error is is: ", max_error(y_IVS, preds))
print("The Mean Absolute Error is: ", mean_absolute_error(y_IVS, preds))
```

# Objective 2

```python
df.loc[(df['total_UPDRS'] <= 40), 'total_UPDRS'] = 1
df.loc[(df['total_UPDRS'] > 40), 'total_UPDRS'] = 0
```

```python
important_columns = ['Jitter(Abs)', 'Shimmer:APQ5', 'Shimmer:APQ11', 'NHR', 'HNR', 'DFA', 'PPE']
X = df.loc[:, df.columns.isin(important_columns)].to_numpy()
y = df['total_UPDRS'].to_numpy()
```

```python
X_train, X_IVS, y_train, y_IVS = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
import numpy as np
from sklearn.model_selection import KFold

kf = KFold(n_splits=20, shuffle=True, random_state=23)
kf.get_n_splits(X_train)

precision_values = []
recall_values = []
f1_values = []
mcc_values = []

# min_samples_leaf
for i in range(1, 21):
    TRUTH_nfold=None
    PREDS_nfold=None
    for train_index, test_index in kf.split(X_train):
        Xk_train, Xk_test = X_train[train_index], X_train[test_index]
        yk_train, yk_test = y_train[train_index], y_train[test_index]

        mdl = DecisionTreeClassifier(min_samples_leaf=i, random_state=3)
        mdl.fit(Xk_train, yk_train)
        preds = mdl.predict(Xk_test)
        if TRUTH_nfold is None:
            PREDS_nfold=preds
            TRUTH_nfold=yk_test
        else:
            PREDS_nfold=np.hstack((PREDS_nfold, preds))
            TRUTH_nfold=np.hstack((TRUTH_nfold, yk_test))

    print(f"Statistics for {i} min_samples_leaf hyperparameter: \n--------------------------")
    print("The Precision is: %7.4f" % precision_score(TRUTH_nfold, PREDS_nfold))
    print("The Recall is: %7.4f" % recall_score(TRUTH_nfold, PREDS_nfold))
    print("The F1 score is: %7.4f" % f1_score(TRUTH_nfold, PREDS_nfold))
    print("The Matthews correlation coefficient is: %7.4f" % matthews_corrcoef(TRUTH_nfold, PREDS_nfold))
    print("--------------------------")
    precision_values.append(precision_score(TRUTH_nfold, PREDS_nfold))
    recall_values.append(recall_score(TRUTH_nfold, PREDS_nfold))
    f1_values.append(f1_score(TRUTH_nfold, PREDS_nfold))
    mcc_values.append(matthews_corrcoef(TRUTH_nfold, PREDS_nfold))
    
# max_depth
for i in range(1, 21):
    TRUTH_nfold=None
    PREDS_nfold=None
    for train_index, test_index in kf.split(X_train):
        Xk_train, Xk_test = X_train[train_index], X_train[test_index]
        yk_train, yk_test = y_train[train_index], y_train[test_index]

        mdl = DecisionTreeClassifier(max_depth=i, random_state=3)
        mdl.fit(Xk_train, yk_train)
        preds = mdl.predict(Xk_test)
        if TRUTH_nfold is None:
            PREDS_nfold=preds
            TRUTH_nfold=yk_test
        else:
            PREDS_nfold=np.hstack((PREDS_nfold, preds))
            TRUTH_nfold=np.hstack((TRUTH_nfold, yk_test))

    print(f"Statistics for {i} max_depth hyperparameter: \n--------------------------")
    print("The Precision is: %7.4f" % precision_score(TRUTH_nfold, PREDS_nfold))
    print("The Recall is: %7.4f" % recall_score(TRUTH_nfold, PREDS_nfold))
    print("The F1 score is: %7.4f" % f1_score(TRUTH_nfold, PREDS_nfold))
    print("The Matthews correlation coefficient is: %7.4f" % matthews_corrcoef(TRUTH_nfold, PREDS_nfold))
    print("--------------------------")
    precision_values.append(precision_score(TRUTH_nfold, PREDS_nfold))
    recall_values.append(recall_score(TRUTH_nfold, PREDS_nfold))
    f1_values.append(f1_score(TRUTH_nfold, PREDS_nfold))
    mcc_values.append(matthews_corrcoef(TRUTH_nfold, PREDS_nfold))
```

```python
fig, plots = plt.subplots(nrows=1, ncols=4, figsize=(30, 10))

# min_samples_leaf
number_of_leaves = [i for i in range(1, 21)]
plots[0].set_xlabel("min_samples_leaf")
plots[0].set_ylabel("Precision")
plots[0].plot(number_of_leaves, precision_values[:20])
plots[1].set_xlabel("min_samples_leaf")
plots[1].set_ylabel("Recall")
plots[1].plot(number_of_leaves, recall_values[:20])
plots[2].set_xlabel("min_samples_leaf")
plots[2].set_ylabel("F1")
plots[2].plot(number_of_leaves, f1_values[:20])
plots[3].set_xlabel("min_samples_leaf")
plots[3].set_ylabel("MCC")
plots[3].plot(number_of_leaves, mcc_values[:20])

fig, plots = plt.subplots(nrows=1, ncols=4, figsize=(30, 10))

# max_depth
plots[0].set_xlabel("max_depth")
plots[0].set_ylabel("Precision")
plots[0].plot(number_of_leaves, precision_values[20:])
plots[1].set_xlabel("max_depth")
plots[1].set_ylabel("Recall")
plots[1].plot(number_of_leaves, recall_values[20:])
plots[2].set_xlabel("max_depth")
plots[2].set_ylabel("F1")
plots[2].plot(number_of_leaves, f1_values[20:])
plots[3].set_xlabel("max_depth")
plots[3].set_ylabel("MCC")
plots[3].plot(number_of_leaves, mcc_values[20:])
```
